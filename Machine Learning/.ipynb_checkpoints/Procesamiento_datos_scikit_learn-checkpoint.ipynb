# Variables categoricas en scikit-learn

Las variables categoricas necesitan un especial tratamiento antes de utilizarlas en cualquier modulo de aprendizaje automatico, la razon es que estos modulos solo trabajan con datos numericos. Por lo anterior es necesario codificar o transformar los datos categoricos en numeros, existen diferentes estrategias para hacerlo, a continuacion estudiaremos cada una de ellas:

## One-hot-encoding

Se asigna el valor de una de las categorias dentro de una matriz, suponga que tiene una variable con los dias de la semana, que puede tomar cualquiera de los siguientes valores:<br>

Sunday-Monday-Tuesday-Wednesday-Thursday-Friday-Saturday

Si requiere representar el dia Monday en el vector, colocara el numero 1 en el indice de Sunday y cero en las demas pocisiones,y asi sucesivamente para cualquier otro valrode dia de la semana que requiera representar. 

![one_hot_encoding](img/one-hot-encoding.png)

Otra manera de codificar variables categoricas es con la simple asignacion de numeros a categorias:<br>
"Estudiantes" es 1, "Profesor" es 2, "Decano" es 3

Ejemplo de one-hot-encoding con scikit-learn

```python
import pandas as pd
titanic = pd.read_csv('https://raw.githubusercontent.com/justmarkham/DAT5/master/data/titanic_train.csv')
```

La librerias pandas ofrece una manera de representar variables categoricas en one-hot-encoding

```python
pd.get_dummies(titanic['Embarked'])
```

Puede adiccionar las varibles creadas al dataset original:

```python
titanic = pd.get_dummies(titanic, columns=['Embarked'])
```

```python
titanic
```

```python
titanic
```

Asignacion de valores a numericos a categorias, sciklit-learn proporciona la siguiente funcionalidad:

```python
from sklearn import preprocessing
titanic = pd.read_csv('https://raw.githubusercontent.com/justmarkham/DAT5/master/data/titanic_train.csv')
```

```python
le = preprocessing.LabelEncoder()
```

```python
titanic['Embarked'] = le.fit_transform(titanic['Embarked'].astype(str))
```

```python
titanic
```

```python
Tambien se puede hacer de una manera pythonista:
```

```python
titanic = pd.read_csv('https://raw.githubusercontent.com/justmarkham/DAT5/master/data/titanic_train.csv')
titanic.Sex.dropna()
titanic.info()
```

## Representación de imagenes

Es comun utilizar imgenes en tareas de machine learning, por lo cual es importante entender como codificarlas en la manera adecuada paea trabajar con los modulos de aprendizaje automatico. Generalmente las imagenes son representadas como una matriz.
Cada celda de esta matriz representa un pixel en particular de la imagen, y cada pixel tiene algun valor basado en el tipo de imagen que es.<br>
Por ejemplo, una imagen a color, cada pixel se representa utilizando tres valores separados, estos valores son RGB donde cada valor es un numero dentro de un rango de 0-255. El color rojo esta representado por 255, 0, 0; el color verde por: 0, 255, 0 y el color azul por: 0, 0, 255.<br>
Las imagenes a color se denominan de tres canales por que cada pixel requiere tres valores para representar su valor; mientras que una imagen a escala de grises solo requiere un canal para representar el valor del pixel, el cual es el grado de intensidad de ese pixel, tipicamente se presenta entre 0 y 1, no obstante tambien puede ser un numero entre 0 y 255, generalmente se divide por 255 para obtener un numero entre 0 y 1; como ejemplo un pixel podria tener un valor de 0,5, para representar la intensidad en una imagen de un solo canal<br>
Las imagenes en escala de grises de un solo canal y las imagenes a color de multiple canal pueden representarse como una matriz tridimensional. Las dos primeras dimensiones representan la altura y el ancho de cada imagen, y la tercera dimension representa el numero de canales.
ejemplo una imagen a escala de grises se representaria como : (6,6,1)<br>
Ejemplo una imagen a color se representaria como : (6,6,1)

Tipicamente en machine learning las imagenes con las que se trabaja son del mismo tamaño, en cuyo caso se utiliza una matriz de cuatro dimensiones para representar una lista de imagenes. Por ejemplo una lista de imagenes a color se representa como: (10, 6, 6, 3).la primera dimension es el numero de imagenes, la segunda y la tercera son la altura y el ancho, y por ultimo son los canales de la imagen.
El tratamiento de imagenes puede llegar a ser muy especializado, en el medio existen multiples algoritmos y librerias para su procesamiento, por ejemplo scikit-learn proporciona una libreria especifica para trabajar con imagenes llamada scikit-image. No obstante utilizaremos la libreria OpenCV para la representacion de imagenes como numero.
